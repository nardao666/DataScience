```
import tensorflow as tf
from sklearn import datasets
```

```
iris = datasets.load_iris()
```

```
X = iris.data
```

```
Y = iris.target
```

```
from sklearn.preprocessing import StandardScaler
```

```
scaler_x = StandardScaler()
```

```
X = scaler_x.fit_transform(X)
```

```
from sklearn.preprocessing import OneHotEncoder
```

```
onehot = OneHotEncoder(categories = 'auto')
```

```
Y = Y.reshape(len(Y), 1)
```

```
Y.shape
```

```
Y = onehot.fit_transform(Y).toarray()
```

```
Y
```

```
from sklearn.model_selection import train_test_split
```

```
X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size = 0.2)
```

```
model = tf.keras.models.Sequential()
```

```
model.add( tf.keras.layers.Dense(units = neuronios_oculta, activation = 'relu' ) )
```

```
model.add( tf.keras.layers.Dropout( 0.1 ) )
```

```
model.add( tf.keras.layers.Dense(units=neuronios_saida, activation = 'softmax' ) )
```

```
model.compile(loss = 'mse', optimizer= 'SGD' , metrics=['binary_accuracy'])
```

```
model.fit(X_treinamento, Y_treinamento, epochs=10000, verbose=0)
```
