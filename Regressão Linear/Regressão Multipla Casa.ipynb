```
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
```

```
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```
base.head()
```

```
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```
base.head()
```

```
X = base.drop('price', axis = 1)
y = base.price
```

```
X.head()
```

```
y.head()
```

```
previsores_colunas = colunas_usadas[1:17]
previsores_colunas
```

```
tf.random.set_seed(42)
```

```
colunas = [tf.feature_column.numeric_column(key = c) for c in previsores_colunas]
```

```
print(colunas[10])
```

```
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```
X_treinamento.shape
```

```
y_treinamento.shape
```

```
X_teste.shape
```

```
def make_input_fn(X, y = None, num_epochs = None, shuffle = True, batch_size = 32):
    def input_function():
        ds = tf.data.Dataset.from_tensor_slices((dict(X), y))
        if shuffle:
            ds = ds.shuffle(1000)
        ds = ds.batch(batch_size).repeat(num_epochs)
        return ds
    return input_function
```

```
ds = tf.data.Dataset.from_tensor_slices((dict(X_treinamento), y_treinamento))
ds
```

```
funcao_treinamento = make_input_fn(X_treinamento, y_treinamento)
```

```
funcao_treinamento()
```

```
funcao_teste = make_input_fn(X_teste, y_teste, num_epochs = 10000, shuffle = False)
```

```
config = tf.estimator.RunConfig(tf_random_seed = 42)
regressor = tf.estimator.LinearRegressor(feature_columns = colunas, config = config)
```

```
regressor.train(input_fn = funcao_treinamento, steps = 10000)
```

```
metricas_treinamento = regressor.evaluate(input_fn = funcao_treinamento, steps = 10000)
```

```
metricas_teste = regressor.evaluate(input_fn = funcao_teste, steps = 10000)
```

```
metricas_treinamento
```

```
metricas_teste
```

```
funcao_previsao = make_input_fn(X_teste, num_epochs = 1, shuffle = False)
```

```
previsoes = regressor.predict(input_fn = funcao_previsao)
```

```
list(previsoes)
```

```
valores_previsoes = []
for p in regressor.predict(input_fn = funcao_previsao):
    valores_previsoes.append(p['predictions'])
```
