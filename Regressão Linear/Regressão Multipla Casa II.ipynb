```
import numpy as np
```

```
def input_train_fn(features, labels, training=True, batch_size=32):
    """An input function for training or evaluating"""
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # Shuffle and repeat if you are in training mode.
    if training:
        dataset = dataset.shuffle(10000)
    
    return dataset.batch(batch_size)
```

```
regressor.train(input_fn = lambda: input_train_fn(X_treinamento, Y_treinamento, training = True), steps= 10000)
```

```
metricas_treinamento = regressor.evaluate(input_fn = lambda: input_train_fn(X_treinamento, Y_treinamento, training = True), steps= 10000)
```

```
metricas_teste = regressor.evaluate(input_fn = lambda: input_train_fn(X_teste, Y_teste, training = False), steps= 10000)
```

```
def input_t_fn(features, training=True, batch_size=32):
    """An input function for training or evaluating"""
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices(dict(features))

    # Shuffle and repeat if you are in training mode.
    if training:
        dataset = dataset.shuffle(10000)
    
    return dataset.batch(batch_size)
```

```
def make_input_fn(data_df, label_df, num_epochs=10, shuffle=True, batch_size=32):
  def input_function():
    ds = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))
    if shuffle:
      ds = ds.shuffle(1000)
    ds = ds.batch(batch_size).repeat(num_epochs)
    return ds
  return input_function
```
